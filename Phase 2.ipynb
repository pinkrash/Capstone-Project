```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('crime',sep=' ')
```

```python
df1.head()
```

```python
df2=pd.read_csv('Crime Type.csv')
df2.head()
```

```python
# From previously collected data
df3=pd.read_csv('IndiaCrimeData.csv')
df3.head()
```

```python
df4=pd.read_csv('total_crime')
df4.head()
```

```python
# making column name same as we have to merge the data
df1['States/UT']=df2['State / UT']
```

```python
df1['Total Crime']=df4['Total Crime']
```

```python
df=pd.merge(df1,df3[['States/UT','Literacy_Rate','Area']],on='States/UT')
df.head()
```

```python
df1.drop(['Sn'],axis=1,inplace=True)
```

# 2.6 Each state crime report.

```python
crime_list=['Murder','Rape','Kidnapping','Crimes Against Children','Extortion','Robbery','Hit & Run','Drugs Trafficking',
           'Illegal Arms']
plt.figure(figsize=(20,25))
plot=1
for i in crime_list:
    plt.subplot(3,3,plot)
    sns.barplot(x=df2['State / UT'],y=df2[i])
    plt.xticks(rotation='vertical')
    plt.title(i)
    plot+=1
plt.show
```
