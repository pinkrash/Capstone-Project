```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('3.8.2.csv')
df.head()
```

```python
df['DISTRICT'].value_counts()
```

```python
filtered_districts = df.groupby(['STATE_UT', 'DISTRICT']).filter(lambda x: len(x) >= 3)
filtered_districts = filtered_districts.sort_values(by='YEAR', ascending=False)
```
