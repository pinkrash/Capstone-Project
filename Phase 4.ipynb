# Data Preperation

```python
x=df[['YEAR', 'MURDER', 'ATTEMPT TO MURDER',
       'CULPABLE HOMICIDE NOT AMOUNTING TO MURDER', 'RAPE', 'CUSTODIAL RAPE',
       'OTHER RAPE', 'KIDNAPPING & ABDUCTION',
       'KIDNAPPING AND ABDUCTION OF OTHERS', 'DACOITY',
       'PREPARATION AND ASSEMBLY FOR DACOITY', 'ROBBERY', 'BURGLARY', 'THEFT',
       'AUTO THEFT', 'OTHER THEFT', 'RIOTS', 'CRIMINAL BREACH OF TRUST',
       'CHEATING', 'COUNTERFIETING', 'ARSON', 'HURT/GREVIOUS HURT',
       'DOWRY DEATHS', 'CAUSING DEATH BY NEGLIGENCE', 'OTHER IPC CRIMES',
       'TOTAL IPC CRIMES', 'total crime against children',
       'total crimes against women', 'total crime against SC',
       'total crime against ST']]
```

# Scaling

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
xscaled = st.fit_transform(x)
DF = pd.DataFrame(xscaled)
```

```python
wcss=[] 
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.title("Elbow Method")
plt.xlabel('No of Cluster')
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state = 42)
kmeans.fit(x)
```

```python
y_kmeans = kmeans.fit_predict(DF)

y_kmeans
```

# 4.1 Creating 3 clusters

```python
cluster = list(y_kmeans)
df['cluster'] = cluster
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x[y_kmeans == 0]['RAPE'], x[y_kmeans == 0]['total crimes against women'], s = 100, c = 'red', label = 'Cluster 1')

plt.scatter(x[y_kmeans == 1]['RAPE'], x[y_kmeans == 1]['total crimes against women'], s = 100, c = 'blue', label = 'Cluster 2')

plt.scatter(x[y_kmeans == 2]['RAPE'], x[y_kmeans == 2]['total crimes against women'], s = 100, c = 'green', label = 'Cluster 3')


plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of Crimes')
plt.xlabel('MURDER')
plt.ylabel('Total crime against Women')
plt.savefig('clusterscatter.jpg')
plt.legend()
plt.show()
```
