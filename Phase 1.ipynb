```python
df.drop(['Rank','Unnamed: 3'],axis=1,inplace=True)
```

```python
# To check the row having literacy rate nan.
nan_rows = df[df['Literacy_Rate'].isna()].index.tolist()
nan_rows
```

```python
# Creating a column just to check increasing rate of population from 2011 to 2023.
df['pop_increas_rate']=((df['population 2023']-df['population  2011'])/df['population  2011'])*100
```

```python
df
```

```python
# Ladakh is 34th row having nan value. I will fill it with the value (after checking increasing rate) of some calculation.
print(df['pop_increas_rate'].mean())
population_of_Ladakh=30000*100/117
print(population_of_Ladakh)
```

```python
# We have observed mean of population increasing rate and calculated population of Ladakh in 2011.
df['population  2011']=df['population  2011'].fillna(population_of_Ladakh)
```
